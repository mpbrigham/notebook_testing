```python
# @title
import numpy as np                 # import numpy
import scipy as sp                 # import scipy
import math                        # import basic math functions
import random                      # import basic random number generator functions
import time

# from IPython import display        

import plotly.graph_objects as go
from ipywidgets import widgets
```

```python
def my_gaussian(x_points, mu, sigma):

    return np.exp(-(x_points-mu)**2/(2*sigma**2))
    
    
def gaussian_update(k=0):

    auditory = my_gaussian(x, auditory_mean, auditory_std)
    visual = my_gaussian(x, visual_mean[k], visual_std)
    posterior_pointwise = auditory * visual
      
    return (
        auditory/np.sum(auditory, keepdims=True),
        visual/np.sum(visual, keepdims=True),
        posterior_pointwise/np.sum(posterior_pointwise, keepdims=True)
    )


def plot_refresh(change):

    auditory, visual, posterior_pointwise = gaussian_update(w_slider.value)

    with w_figure.batch_update():
        w_figure.data[0].y = auditory
        w_figure.data[1].y = visual
        w_figure.data[2].y = posterior_pointwise
```

```python
x = np.arange(-10, 11, 0.1)

visual_mean = np.arange(-8,9,0.2)
visual_std  = 1

auditory_mean = 0
auditory_std  = 1.5

w_slider = widgets.IntSlider(
    value=1.0,
    max=visual_mean.size,
    step=1.0,
    description='Step:'
)

w_slider.observe(plot_refresh, names='value')

auditory, visual, posterior_pointwise = gaussian_update()

trace1 = go.Scatter(x=x, y=auditory, opacity=0.9, name='Auditory')
trace2 = go.Scatter(x=x, y=visual,  opacity=0.9, name='Visual')
trace3 = go.Scatter(x=x, y=posterior_pointwise,  opacity=0.9, name='Posterior')

layout = {
    'title': 'My dynamic plot',
    'xaxis': {'title': 'Probability'},
    'yaxis': {'title': 'Orientation (Degrees)'},
}

w_figure = go.FigureWidget(
    data=[trace1, trace2, trace3],
    layout=layout
)

# display(widgets.VBox([w_slider, w_figure]))

display(w_slider)
display(w_figure)
```

```python
w_figure = go.FigureWidget(
    data=[trace1, trace2, trace3],
    layout=layout
)

display(w_figure)

for k in range(5):

    time.sleep(1)
    with w_figure.batch_update():
        w_figure.data[1].y = 2*visual
```
